```python
import json

T = CartanType(['A', 5])
W = WeylGroup(T,prefix="s")
[s1, s2, s3, s4, s5] = W.simple_reflections()
KLP = json.load( open( "kl_polys/a3.json" ) )

WCR = WeylCharacterRing(T, style="coroots", base_ring=QQ)
R = WeightRing(WCR)
# EXPI MODIFIED, THIS CODE ONLY WORKS IN TYPE A
f = [w.coerce_to_sl() for w in WCR.fundamental_weights()]

n = len(W.simple_reflections())
e = s1^2
w0 = W.long_element()
S.<q> = LaurentPolynomialRing(QQ)
KL = KazhdanLusztigPolynomial(W,q)
q = 11

def Iw(w):
    a = []
    for i in range(1, n+1):
        if w.has_right_descent(i):
            a.append(i)
    return tuple(a)

def RSK_of_perm(w):
    m = matrix(w).rows()
    n1 = []
    for row in m:
        new_row = []
        for j in row:
            new_row.append(int(j))
        n1.append(new_row)
    return RSK(n1)

def tableau_type(l):
    lengths = []
    for r in l[0]:
        lengths.append(len(r))
    return lengths

def cell(w):
    return tableau_type(RSK_of_perm(w))

invols = []
for w in W:
    if w*w == e:
        invols.append(w)

cell_dict = {}
for y in W:
    c = tuple(cell(y))
    if c not in cell_dict:
        cell_dict[c] = [y]
    elif c in cell_dict:
        cell_dict[c].append(y)

d = cell_dict

check_repeats = {}
for c in d:
    print(c)
    check_repeats[c] = {}
    for b in d[c]:
        if b*b == s1*s1:
            if Iw(b) not in check_repeats[c]:
                check_repeats[c][Iw(b)] = [b]
            else:
                check_repeats[c][Iw(b)].append(b)
    for lis in check_repeats[c]:
        if len(check_repeats[c][lis]) != 1:
            print("HELLO AT LALLALALALALAALLALALALLALALALALLALALALALALLALL!!!!!!!")
        print(len(check_repeats[c][lis]), list(lis), check_repeats[c][lis])
    print()
```

```python
def build_matrix():
    m = []
    for i in range(len(W)):
        print(i + 1, "out of", len(W), end='\r')
        r = []
        for j in range(len(W)):
            r.append(woke_pairing(d_sub[i], d_sup[j]))
        m.append(r)
    return m
bigmat = build_matrix()
#bigmat[i][j] = woke_pairing(d_sub[i], d_sup[j]) for all i, j
alls = list(range(len(W)))
pops = list(range(len(W)))
order = []
while len(pops) > 0:
    for a in pops:
        good = True
        for b in pops:
            if b != a and bigmat[b][a] != 0:
                good = False
        if good:
            order.append(a)
            pops.remove(a)

order.reverse()
print("Computed the matrix of pairings and the upper triangular order.")
```

```python
def lc(i):
    a = bigmat[i][i]
    d = dict(WCR(a))
    if len(d) != 1:
        print("LC ERROR")
    else:
        return list(d.values())[0]

counter = 0
d_dual = {}
for j in order:
    counter += 1
    print(counter, "out of", len(W), end='\r')
    d_dual[j] = d_sup[j]/lc(j)
    for i in order:
        if  i != j and woke_pairing(d_sub[i], d_dual[j]) != 0:
            d_dual[j] = d_dual[j] - woke_pairing(d_sub[i], d_dual[j])*d_sup[i]/lc(i)
print("Computed the dual basis as the dictionary d_dual.")
```

```python
#OPTIONAL CELL: CHECKING THE DUAL BASIS IS VALID.
valid = True
for i in range(len(W)):
    print(i + 1, "out of", len(W), end='\r')
    for j in range(len(W)):
        a = woke_pairing(d_sub[i], d_dual[j])
        if a != 0:
            if i != j or a != 1:
                print("ISSUE AT", i, j)
                valid = False
if valid:
    print("Checked dual basis, all is good.")
else:
    print("ISSUES! Dual basis is NOT correct.")
```
