```python
rank = 3  # Change this number only

T = CartanType(['B', rank])
W = WeylGroup(T, prefix="s")
s = W.simple_reflections()
for i in range(1, rank + 1):
    globals()[f"s{i}"] = s[i]

WCR = WeylCharacterRing(T, style="coroots", base_ring=QQ)
R = WeightRing(WCR)
f = [w.coerce_to_sl() for w in WCR.fundamental_weights()]
w0 = W.long_element()

def Iw(w):
    return [i for i in range(1, rank + 1) if w.has_right_descent(i)]

def expI(indices):
    a = R.one()
    for i in indices:
        a *= R(f[i - 1])
    return a

def esub(w):
    return expI(Iw(w)).weyl_group_action(w)

def Iweight(w):
    return sum(f[i - 1] for i in Iw(w))

rho = Iweight(w0)

def woke_weightpairing(l):
    a = WCR.dot_reduce(l - rho)
    return 0 if a[0] == 0 else a[0] * WCR(a[1])

def inprod(a, b):
    return sum(da * db * v.inner_product(u)
               for v, da in dict(a).items()
               for u, db in dict(b).items())

def rootprod(a, b):
    return 2 * inprod(a, b) / inprod(b, b)

def dom(a):
    return all(rootprod(a, R(b)) >= 0 for b in R.simple_roots())

def woke_pairing(a, b):
    return sum(coeff * woke_weightpairing(wt)
               for wt, coeff in dict(R(a * b)).items())

def hom(a, b):
    return woke_pairing(b, R(rho) * a.scale(-1))

def wt(a):
    #Takes in some R(1,1) and spits out the raw weight, i.e. defined by R(wt(R(a,b))) = R(a,b).
    return list((dict(a)).keys())[0]

def pchar(a):
    domele_w_pairs = [
        (a.weyl_group_action(y), y) for y in W if dom(a.weyl_group_action(y))
    ]
    domele, w = min(domele_w_pairs, key=lambda pair: pair[1].length())
    w_inv = w.inverse()
    redword = w_inv.reduced_word()
    return WCR.demazure_character(wt(domele), word=redword)

def p(w):
    return pchar(esub(w).scale(-1)).scale(-1)

def check_dual(d1, d2):
    for w in W:
        for y in W:
            a = woke_pairing(d1[w], d2[y])
            if (w != y and a != 0) or (w == y and a != 1):
                print("Fails at", w, y)
    print("Checked.")

Wo = sorted(W, key=lambda w: w.length())
print("Initialized setup. Weyl group has size", str(len(W)) + ".")
```
