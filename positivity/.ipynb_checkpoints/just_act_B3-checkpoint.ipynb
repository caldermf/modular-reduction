```python
l = ['B',3]
T = CartanType(l)
laff = l.copy()
laff.append(1)
W = WeylGroup(CartanType(l),prefix="s")
[s1, s2, s3] = W.simple_reflections()
S = W.simple_reflections()
w0 = W.long_element()
lw0 = w0.length()
p = 29
e = s0*s0

WCR = WeylCharacterRing(T, style="coroots", base_ring=QQ)
R = WeightRing(WCR)
f = [w.coerce_to_sl() for w in WCR.fundamental_weights()]

lr = R(2,0)

r = list(W.reflections())[-1]
#alternatively S[0] I think works!

def inprod(a, b):
    da = dict(a)
    db = dict(b)
    t = 0
    for v in da:
        for u in db:
            t += da[v] * db[u] * v.inner_product(u)
    return t

def rootprod(a, b):
    return 2*inprod(a,b)/inprod(b,b)

def s0act(a):
    c = p/2 - rootprod(a, lr)
    if c == 0:
        return a
    else:
        return (a * R(lr).scale(c))

def act(l, a):
    l.reverse()
    for i in l:
        if i == 0:
            a = s0act(a)
        else:
            a = a.weyl_group_action(S[i])
    return a

def revact(l, a):
    for i in l:
        if i == 0:
            a = s0act(a)
        else:
            a = a.weyl_group_action(S[i])
    return a

def in_domain(a):
    if rootprod(a, lr) < 0 or rootprod(a, lr) > p/2:
        return False
    else:
        for b in R.simple_roots():
            if rootprod(a, R(b)) < 0:
                return False
        return True
#Ah ok. Last issue is that s0 doesn't square to 0...
```
