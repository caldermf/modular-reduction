```python
l = ['B',2]
T = CartanType(l)
laff = l.copy()
laff.append(1)
W = WeylGroup(CartanType(l),prefix="s")
[s1, s2] = W.simple_reflections()
S = W.simple_reflections()
Waff = WeylGroup(CartanType(laff),prefix="s")
[s0, s1, s2] = Waff.simple_reflections()
Saff = Waff.simple_reflections()
w0 = W.long_element()
lw0 = w0.length()
p = 3
e = s0*s0

WCR = WeylCharacterRing(CartanType(l), style="coroots", base_ring=QQ)
R = WeightRing(WCR)
b3 = R
B3 = WCR
f = [w.coerce_to_sl() for w in WCR.fundamental_weights()]

r = list(W.reflections())[-1]
#alternatively S[0] I think works!

def inprod(a, b):
    da = dict(a)
    db = dict(b)
    t = 0
    for v in da:
        for u in db:
            t += da[v] * db[u] * v.inner_product(u)
    return t

def rootprod(a, b):
    return 2*inprod(a,b)/inprod(b,b)
    
lr = R(2,0,0)

def s0act(a):
    c = p/2 - rootprod(a, lr)
    if c == 0:
        return a
    else:
        return (a * R(lr).scale(c))

def act(w, a):
    l = w.reduced_word()
    l.reverse()
    for i in l:
        print(a)
        if i == 0:
            a = s0act(a)
        else:
            a = a.weyl_group_action(S[i])
    return a

def in_domain(a):
    if rootprod(a, lr) < 0 or inprod(a, lr) > p:
        return False
    else:
        for b in R.simple_roots():
            if rootprod(a, R(b)) < 0 or rootprod(a, R(b)) >= p:
                return False
        return True

def find_refl(a, y=s0*Waff(w0)*s0*Waff(w0)*s0*Waff(w0)):
    interv = Waff.bruhat_interval(e, y)
    poss = []
    for w in interv:
        b = act(w, a)
        if in_domain(b):
            poss.append((w.inverse(), b))
    if len(poss) == 0:
        return False
    minimal = poss[0]
    for z in poss:
        if z[0].length() < minimal[0].length():
            minimal = z
    return minimal
#Ah ok. Last issue is that s0 doesn't square to 0...
```

```python
def manual_weight_dict(a):
    d = {}
    l = dict(a)
    for k in l:
        d[R(k)] = l[k]
    return d

def weight_dict(w):
    d = {}
    l = dict(Mw[w])
    for k in l:
        d[R(k)] = l[k]
    return d

rho = R.one()
for wt in f:
    rho = rho * R(wt)

negrho = R.one()
for wt in f:
    negrho = negrho * R(wt).scale(-1)



def translate_dict(d):
    #give me a weight dict I'll translate each of the weights
    new_d = {}
    for k in d:
        j = find_refl(k*rho)
        if j == False:
            return "ERROR"
        else:
            new_d[k] = (Waff(j[0]), j[1])
    return new_d

def all_weights(d):
    weights = []
    for k in d:
        weights.append(d[k][1])
    return(weights)

def all_weyls(d):
    weyls = []
    for k in d:
        weyls.append(d[k][0])
    return(weyls)

invols = []
for w in W:
    if w * w == W(s1*s1) and w != W(s1*s1):
        invols.append(w)
```

```python
w0 = Waff(w0)
LRing.<q> = LaurentPolynomialRing(QQ)
KL = KazhdanLusztigPolynomial(Waff,q)

def mu(x, y):
    a = KL.P(x, y)
    if 2*(a.degree()) == y.length() - x.length() - 1:
        return a.coefficients()[-1]
    else:
        return 0

def can_appear(z, r):
    sr = Saff[r]
    if (w0*z).length() != z.length() - lw0:
        return False
    elif (z*sr).length() > z.length():
        return False
    else:
        return True

def guys_appearing(a, r):
    #guys which appear in the expansion of C_{a}*T_r with nontrivial mu coeff, assuming lengths all add
    #ofc C_{a*r} and C_{a} appear but aren't listed!
    l = []
    for z in W.bruhat_interval(s1*s1, a):
        if can_appear(z, r):
            m = mu(z, a)
            if m != 0:
                l.append((z, m))
    return l

def mult(list_a, r_index):
    a = e
    for i in list_a:
        a = a * Saff[i]
    r = Saff[r_index]
    total = C[w0_list + list_a] + C[w0_list + list_a + [r_index]]
    for guy in guys_appearing(a, r):
        total += guy[1] * C[guy[0].reduced_word()]
    return total

def single_mult(a, r):
    sr = Saff[r]
    if (w0*a).length() != a.length() - lw0:
        print("ERROR")
        return "ERROR"
    elif (a*sr).length() < a.length():
        d = {a : -1}
    else:
        d = {a : 1, a*sr : 1}
        for guy in guys_appearing(a, r):
            d[guy[0]] = guy[1]
    return d

def multi_mult(d, r):
    new_d = {}
    for w in d:
        dw = single_mult(w, r)
        for y in dw:
            if y not in new_d:
                new_d[y] = dw[y]
            else:
                new_d[y] += dw[y]
    nonzero_d = {}
    for key in new_d:
        if new_d[key] != 0:
            nonzero_d[key] = new_d[key]
    return nonzero_d

def full_mult(a, l):
    if len(l) == 0:
        return {a : 1}
    first = l.pop(0)
    d = single_mult(a, first)
    while len(l) > 0:
        new_r = l.pop(0)
        #print("after " + str(new_r) + ":")
        d = multi_mult(d, new_r)
    return d

def create_nondom_vdict(w):
    we = fulldict[W(w)]
    #we = weights and elements
    total = {}
    for k in we:
        y = we[k][0]
        a = we[k][1]
        if type(y) == type(1):
            guys = full_mult(Waff(w0), [])
        else:
            guys = full_mult(Waff(w0), y.reduced_word())
        for g in guys:
            mu = act(w0*g, a)
            if mu not in total:
                total[mu] = guys[g] * weight_dict(Waff(w))[k]
            else:
                total[mu] = total[mu] + guys[g] * weight_dict(Waff(w))[k]
    return total

def purge_nondom_and_shift(d):
    new_d = {}
    for k in d:
        kisgood = True
        shiftedk = k * rho.scale(-1)
        for b in R.simple_roots():
            if inprod(R(b), shiftedk) < 0:
                kisgood = False
        if kisgood:
            new_d[shiftedk] = d[k]
    return new_d

def create_vdict(w):
    return purge_nondom_and_shift(create_nondom_vdict(w))
```
