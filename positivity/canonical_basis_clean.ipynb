```python
l = ['A', 4]
laff = l.copy()
laff.append(1)
Wfin = WeylGroup(CartanType(l),prefix="s")
W = WeylGroup(CartanType(laff),prefix="s")
[s0, s1, s2, s3, s4] = W.simple_reflections()
S = W.simple_reflections()
w0 = W(Wfin.long_element())
lw0 = w0.length()
p = 7
e = s0*s0
LRing.<q> = LaurentPolynomialRing(QQ)
KL = KazhdanLusztigPolynomial(W,q)

def mu(x, y):
    a = KL.P(x, y)
    if 2*(a.degree()) == y.length() - x.length() - 1:
        return a.coefficients()[-1]
    else:
        return 0

def can_appear(z, r):
    if (w0*z).length() != z.length() - lw0:
        return False
    elif (z*r).length() > z.length():
        return False
    else:
        return True

def guys_appearing(a, r):
    #guys which appear in the expansion of C_{w0*a}*T_r with nontrivial mu coeff, assuming lengths all add
    #ofc C_{w0*a*r} and C_{w0*a} appear but aren't listed!
    l = []
    for z in W.bruhat_interval(w0, w0*a):
        if can_appear(z, r):
            m = mu(z, w0*a)
            if m != 0:
                l.append((z, m))
    return l

def mult(list_a, r_index):
    a = e
    for i in list_a:
        a = a * S[i]
    r = S[r_index]
    total = C[w0_list + list_a] + C[w0_list + list_a + [r_index]]
    for guy in guys_appearing(a, r):
        total += guy[1] * C[guy[0].reduced_word()]
    return total

def single_mult(a, r):
    sr = S[r]
    if (w0*a).length() != a.length() - lw0:
        return "ERROR"
    elif (a*sr).length() < a.length():
        d = {a : -1}
    else:
        d = {a : 1, a*sr : 1}
        for guy in guys_appearing(w0*a, sr):
            d[guy[0]] = guy[1]
    return d

def multi_mult(d, r):
    new_d = {}
    for w in d:
        dw = single_mult(w, r)
        for y in dw:
            if y not in new_d:
                new_d[y] = dw[y]
            else:
                new_d[y] += dw[y]
    nonzero_d = {}
    for key in new_d:
        if new_d[key] != 0:
            nonzero_d[key] = new_d[key]
    return nonzero_d

def full_mult(a, l):
    first = l.pop(0)
    d = single_mult(a, first)
    while len(l) > 0:
        new_r = l.pop(0)
        #print("after " + str(new_r) + ":")
        d = multi_mult(d, new_r)
    return d

def multi_full_mult(d, l):
    while len(l) > 0:
        new_r = l.pop(0)
        d = multi_mult(d, new_r)
    return d
```
